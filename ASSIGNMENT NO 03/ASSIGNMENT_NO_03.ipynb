```python
import csv

# Open the CSV file in read mode
with open('/content/sample_data/cofee .csv', 'r') as file:
    # Create a CSV reader object
    reader = csv.DictReader(file)

    # Initialize an empty list to store the data
    data = []

    # Iterate over each row in the CSV file
    for row in reader:
        # Append the row as a dictionary to the data list
        data.append(dict(row))

# Print the contents of the data list
for row in data:
    print(row)

import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/sample_data/cofee .csv')

# Convert DataFrame to matrix
matrix = df.to_numpy()

# Perform matrix operations

# Transpose the matrix
transposed_matrix = np.transpose(matrix)

# Print the results
print("Original Matrix:")
print(matrix)
print("Transposed Matrix:")
print(transposed_matrix)

# Split the matrix into separate arrays
array1 = matrix[:, :6]  # Select the first 6 columns
array2 = matrix[:, 6:]  # Select the remaining columns

# Horizontally stack the arrays
horizontal_stack = np.hstack((array1, array2))

# Vertically stack the arrays
vertical_stack = np.vstack((array1, array2))

# Print the results
print("Horizontal Stack:")
print(horizontal_stack)
print("Vertical Stack:")
print(vertical_stack)

# Generate a custom sequence
custom_sequence = ""

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    country = row['Country of Origin']
    bags = str(row['Number of Bags'])
    weight = row['Bag Weight']
    aroma = str(row['Aroma'])
    flavor = str(row['Flavor'])
    aftertaste = str(row['Aftertaste'])
    acidity = str(row['Acidity'])
    balance = str(row['Balance'])
    sweetness = str(row['Sweetness'])
    moisture = str(row['Moisture Percentage'])
    color = row['Color']
    defects = str(row['Category Two Defects'])

    # Create a custom sequence based on the row data
    sequence = f"Country: {country}, Bags: {bags}, Weight: {weight}, Aroma: {aroma}, Flavor: {flavor}, Aftertaste: {aftertaste}, Acidity: {acidity}, Balance: {balance}, Sweetness: {sweetness}, Moisture: {moisture}, Color: {color}, Defects: {defects}\n"

    # Append the sequence to the custom_sequence string
    custom_sequence += sequence

# Print the custom sequence
print(custom_sequence)

# Arithmetic Operations
# Calculate the sum of 'Number of Bags' column
sum_bags = df['Number of Bags'].sum()

# Statistical Operations
# Calculate the minimum value of 'Aroma' column
min_aroma = df['Aroma'].min()

# Calculate the maximum value of 'Flavor' column
max_flavor = df['Flavor'].max()

# Print the results
print("Arithmetic Operations:")
print("Sum of Number of Bags:", sum_bags)
print()
print("Statistical Operations:")
print("Minimum Aroma:", min_aroma)
print("Maximum Flavor:", max_flavor)
print()

# Copy the arrays
country_of_origin = df['Country of Origin'].to_numpy()
number_of_bags = df['Number of Bags'].to_numpy()
bag_weight = df['Bag Weight'].to_numpy()
aroma = df['Aroma'].to_numpy()
flavor = df['Flavor'].to_numpy()
aftertaste = df['Aftertaste'].to_numpy()
acidity = df['Acidity'].to_numpy()
balance = df['Balance'].to_numpy()
sweetness = df['Sweetness'].to_numpy()
moisture_percentage = df['Moisture Percentage'].to_numpy()
color = df['Color'].to_numpy()
category_two_defects = df['Category Two Defects'].to_numpy()

# View the arrays
print("Country of Origin:")
print(country_of_origin)
print("Number of Bags:")
print(number_of_bags)
print("Bag Weight:")
print(bag_weight)
print("Aroma:")
print(aroma)
print("Flavor:")
print(flavor)
print("Aftertaste:")
print(aftertaste)
print("Acidity:")
print(acidity)
print("Balance:")
print(balance)
print("Sweetness:")
print(sweetness)
print("Moisture Percentage:")
print(moisture_percentage)
print("Color:")
print(color)
print("Category Two Defects:")
print(category_two_defects)

# Stacking (concatenating vertically) two DataFrames
stacked_df = pd.concat([df, df], axis=0)

# Searching for rows where the 'Country of Origin' is 'Taiwan'
taiwan_df = df[df['Country of Origin'] == 'Taiwan']

# Sorting the DataFrame by 'Number of Bags' in descending order
sorted_df = df.sort_values(by='Number of Bags', ascending=False)

# Counting the occurrences of each unique value in the 'Color' column
color_counts = df['Color'].value_counts()

# Broadcasting a new value to all rows in the 'Sweetness' column
df['Sweetness'] = 10.0

# Printing the results
print("Stacked DataFrame:")
print(stacked_df)
print("Taiwan DataFrame:")
print(taiwan_df)
print("Sorted DataFrame:")
print(sorted_df)
print("Color Counts:")
print(color_counts)
print("Broadcasted DataFrame:")
print(df)

```
